# Own Trained Dataset with Doc2Vec
### 1. Data import
### 2. Generate and train the model
### 3. Infer vectors
### 4. Create dataframe, add score and save

**1. Data import**

```
from google.colab import drive
drive.mount('/content/drive')
```

```
CSVLInk = r'/content/drive/My Drive/DM_Project - Preprocessing/test2_token_excludeSpecPuct.csv'
```

```
import pandas as pd
import io
import time

hotelData = pd.read_csv(CSVLInk) 
review = hotelData["Review"].apply(lambda row: row.strip("']['").split("', '"))
score = hotelData["Reviewer_Score"]
```

**2. Generate and train the model**

```
from gensim.test.utils import common_texts
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
firstTime = time.time()

documents = [TaggedDocument(doc, [i]) for i, doc in enumerate(review.to_list())]
model = Doc2Vec(documents, vector_size=300, window=15,dm=0, min_count=1,epochs=400,workers=-1)
print ("--- %s seconds ---" % round(time.time()-firstTime,4))
```

**3. Infer vectors**

```
firstTime = time.time()

# Inference hyper-parameters # Sorce Paper: https://arxiv.org/pdf/1607.05368.pdf?source=post_page---------------------------  
start_alpha=0.01
min_alpha = 0.0001
infer_epoch=750

vectors = []
for text in review.to_list():
    vectors.append(model.infer_vector(text, alpha= start_alpha, min_alpha= min_alpha, epochs= infer_epoch))
print ("--- %s seconds ---" % round(time.time()-firstTime,4))
```

**4. Create dataframe, add score and save**

```
df = pd.DataFrame(vectors)
df["Reviewer_Score"] = score
```

```
df.to_csv("/content/drive/My Drive/Owntrained.csv", index=False)
```
